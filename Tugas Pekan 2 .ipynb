### **4. Other Findings**

```python
# Regression 'battery_power' & 'ram'
sns.jointplot(x='battery_power', y='ram', data=df, kind='reg')
```

> Berdasarkan distribusi data bisa dilihat tidak ada pola korelasi yang terbentuk 

```python
# Join Plot 'price_class' & 'ram'
sns.jointplot(x='price_class', y='ram', data=df, kind='hex')
```

> Disini bisa dilihat bahwa ternyata ada HP dengan kapasitas RAM moderat (cukup) yang berada pada 2 kategori harga (Mahal & Murah). **Apabila pertimbangan membeli lebih mengutamakan kapasitas RAM maka HP dengan kapasitas RAM tertinggi (2000 - 2500 MB) yang berada pada price_class 0 (murah) adalah the good deal**

```python
sns.jointplot(x='price_class', y='pc', data=df, kind='kde')
```

> Disini bisa dilihat bahwa garis kernel density pada HP mahal (1) sebagian besar berada pada tingkat resolusi 6-11 Megapixel yang mana nilai resolusi tersebut sangat banyak didapati pada HP dengan kategori murah (0). **Jadi hampir tidak ada perbedaan HP mahal atau murah dari sisi resolusi kamera**

### **4. Data Binning**

```python
sns.displot(df.ram, kde=True, bins=100)
```

```python
# Data Binning
df_bin = df[['price_class', 'ram']]
df_bin['ram_category'] = pd.cut(df_bin.ram, bins=[0, 1500, 3000, 4000], labels=['Low Performance', 'Medium Performance', 'High Performance'])
df_bin
```

```python
# Visualize ['ram_category']
df_bin.ram_category.value_counts().plot(kind='bar', title='RAM by Category')
```
